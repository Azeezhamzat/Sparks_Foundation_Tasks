```python
# Performing the bivariate analysis
```

Analyzing the two numeric variables from a dataset is known as numeric-numeric analysis. We can analyze it in three different ways.

```python
plt.scatter(dataset.Profit,dataset.Discount,marker = "*")

plt.xlabel("Profit")
plt.ylabel("Discount")
plt.title("Profit vs Discount")
plt.show()
```

```python
#plot the scatter plot of balance and age variable in data
dataset.plot.scatter(x="Sales",y="Profit")
plt.xlabel("Profit")
plt.ylabel("Sales")
plt.title("Profit vs Sales")
plt.show()
```

```python
# plotting the pair plots
sns.pairplot(dataset,hue='Category',vars=['Sales','Quantity','Discount','Profit'])
```

```python
# In Sales the maximum profit is in the category of Technology than in Office Supplies 
# The demand is also of Technology related goods and Office Supplies
# In the discount in technology as the discount increases the profit is quite low.
# Technology and office supplies are having the maximum profit in terms of the quantity sold.
```

```python
# plotting the pair plots
sns.pairplot(dataset,hue='Sub-Category',vars=['Profit','Discount'])
```

```python
# At low discount maximum profit can  be scene in the Machines and Copiers
```
